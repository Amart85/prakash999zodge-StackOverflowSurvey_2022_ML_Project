#Elbo method

```python
from sklearn.cluster import KMeans
```

```python
wcss = []
for c in range(1,11):
  km = KMeans(n_clusters = c, init = 'k-means++', random_state = 0)
  km.fit(x1)
  wcss.append(km.inertia_)
```

```python
wcss
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(range(1,11),wcss)
```

```python
km = KMeans(n_clusters = 2, random_state = 42)
```

```python
km.fit(x1)
```

```python
# km.predict(x1) returns the cluster number which belongs to some record
identified_cluster = km.predict(x1)
```

```python
# returns centroids
km.cluster_centers_ 
```

```python
# creating new table for new cluster column
data_with_cluster = x1.copy()
```

```python
data_with_cluster['Cluster'] = identified_cluster
```

```python
data_with_cluster
```

```python
data_with_cluster.Cluster.value_counts()
```

```python
# for our understanding lets see the coordinates of two centroids
cluster_center = km.cluster_centers_
temp = pd.DataFrame(cluster_center)
temp
```

```python
c1x = km.cluster_centers_[0][0]
c1y = km.cluster_centers_[0][1]
c2x = km.cluster_centers_[1][0]
c2y = km.cluster_centers_[1][1]
```

```python
plt.scatter(data_with_cluster['YearsCode'],data_with_cluster['WorkExp'],c = data_with_cluster['Cluster'])
plt.scatter([c1x,c2x],[c1y,c2y], c = 'green')
plt.xlabel('Years Code')
plt.ylabel('Work Experience')
plt.title('KMeans Clustering')
plt.show()
```
